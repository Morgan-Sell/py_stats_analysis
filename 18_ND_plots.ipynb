# ND Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_original = pd.read_csv("data/Diabetes.csv")

df_original.head()
```

```python
cols = [c for c in df_original.columns if c not in ["Pregnacies", "Outcome"]]
cols
```

```python
df = df_original.copy()
df[cols] = df[cols].replace({0: np.NaN})

df.head()
```

```python
df.describe()
```

--
## Scatter Matrix

```python
pd.plotting.scatter_matrix(df, figsize=(12,12));
```

```python
df2 = df.dropna()
colors = df2.Outcome.map(lambda x: "#44d9ff" if x else "#f95b4a")

pd.plotting.scatter_matrix(df2, figsize=(12, 12), color=colors);
```

****
## Correlation Plots

```python
df.corr()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.heatmap(df.corr(), ax=ax);
```

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.heatmap(df.corr(), annot=True, cmap="viridis", fmt="0.2f", ax=ax);
```
